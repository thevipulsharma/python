### Different ways to create dataframe
1. Using csv
2. Using excel
3. From python dictionaries
4. From list of tuples
5. From list of dictionaries

```python
import pandas as pd
csv_df = pd.read_csv("data/weather_data.csv")
csv_df
```

```python
excel_df = pd.read_excel("data/weather_data.xlsx", "Sheet1")
excel_df
```

```python
weather_data = {
    "day": ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017'],
    "temperature": [32, 35, 28, 24, 32, 31],
    "windspeed": [6, 7, 2, 7, 4, 2],
    "event": ["Rain", "Sunny", "Snow", "Snow", "Rain", "Sunny"]
}
dict_df = pd.DataFrame(weather_data)
dict_df
```

```python
weather_data = [
    ("1/1/2017", "Rain", 32, 6),
    ("1/2/2017", "Sunny", 35, 7)
]

tuple_df = pd.DataFrame(weather_data, columns=["day", "event", "temperature", "windspeed"])
tuple_df
```

```python
weather_data = [
    {"day": "1/1/2017", "event":"Rain", "temperature":32, "windspeed":6},
    {"day": "1/2/2017", "event":"Sunny", "temperature":35, "windspeed":7}
]

diclist_df = pd.DataFrame(weather_data)
diclist_df
```
