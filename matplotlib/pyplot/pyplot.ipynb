### Working with multiple figures and axes

```python
import matplotlib.pyplot as plt
plt.figure(1)
plt.subplot(211) # Identical to plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4])
plt.subplot(212)
plt.plot([11, 22, 33, 44])
plt.show()
```

### Working with texts

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5,6,7],[11,55,33,44,99,22,66])
plt.title('Plot Title')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.text(2, 88, 'text-1\ntext-2')
plt.grid(True)
plt.show()
```
